```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
def d2_gauss(x,y):
    sigma = 3
    return (1/(2*np.pi*sigma))*np.exp((-x**2-y**2)/(2*sigma**2))
```

```python
x_rng = np.linspace(-5.0,5.0,300)
y_rng = np.linspace(-5.0,5.0,300)
array = np.zeros([300,300])

X, Y = np.meshgrid(x_rng, y_rng)
Z = d2_gauss(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='BuPu')

plt.show()
```
